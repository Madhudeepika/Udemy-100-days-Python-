Data Frames

```python
data = {'Name': ['Amsu', 'Madhu'], 'Age': [25, 30]}
df = pd.DataFrame(data)
```

Viewing and Inspecting Data

```python
df.head()          # First 5 rows
df.tail(3)         # Last 3 rows
df.info()          # Summary of the DataFrame
df.describe()      # Stats summary
df.shape           # (rows, columns)
```

Selecting Data

```python
df['Name']          # Column by name
df[['Name', 'Age']] # Multiple columns
df.loc[0]           # Row by label
df.iloc[1]          # Row by index
df.iloc[0:2, 0:1]   # Slice rows and columns
```

Data Cleaning

```python
df.isnull().sum()            # Check for missing
df.dropna()                  # Drop missing
df.fillna(0)                 # Fill missing with 0
df.duplicated().sum()        # Check duplicates
df.drop_duplicates()         # Remove duplicates
```

Data Manipulation

```python
df['Age'] = df['Age'] + 1        # Modify column
df['Salary'] = df['Age'] * 2     # Add new column
df.rename(columns={'Name': 'FullName'}, inplace=True)
df.sort_values(by='Age', ascending=False)
```

Aggregation and grouping

```python
df.groupby('Age').count()
df.groupby('Age')['NewCol'].mean()
df['Age'].value_counts()
```

Merging and Joining

```python
import pandas as pd

# Sample DataFrames
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Amsu', 'Madhu', 'Deepika']})
df2 = pd.DataFrame({'ID': [2, 3, 4], 'Marks': [85, 90, 75]})

# Merge on 'ID' (only matching IDs will be kept)
result = pd.merge(df1, df2, on='ID', how='inner')

print(result)

```
